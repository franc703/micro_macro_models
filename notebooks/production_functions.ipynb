```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to calculate CES expenditure share of capital
def ces_expenditure_share_capital(p, alpha, sigma):
    return ((1 - alpha) * (p ** (sigma - 1))) / (alpha + (1 - alpha) * (p ** (sigma - 1)))

# Function to calculate Cobb-Douglas expenditure share of capital
def cd_expenditure_share_capital(alpha):
    return alpha

# Generate a range of prices
prices = np.linspace(0.8, 1.2, 100)

# Calculate expenditure shares for CES production function
ces_expenditure_shares = [ces_expenditure_share_capital(p, 0.6, 0.8) for p in prices]

# Calculate expenditure shares for Cobb-Douglas production function
cd_expenditure_shares = [cd_expenditure_share_capital(0.6) for _ in prices]

# Print the output of the functions
print("Expenditure shares for CES production function:")
for p, share in zip(prices, ces_expenditure_shares):
    print(f"Price: {p:.2f}, Share of capital: {share:.2f}")

print("\nExpenditure shares for Cobb-Douglas production function:")
for p, share in zip(prices, cd_expenditure_shares):
    print(f"Price: {p:.2f}, Share of capital: {share:.2f}")

# Plot the results
plt.plot(prices, ces_expenditure_shares, label='CES')
plt.plot(prices, cd_expenditure_shares, label='Cobb-Douglas')
plt.xlabel('Price')
plt.ylabel('Expenditure Share')
plt.legend()
plt.show()
```
